# Decision Tree Classifiers
[Documentation](https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html)


```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn import metrics
import matplotlib.pyplot as plt
```

```python
# import data
iris = datasets.load_iris()

X = iris.data[:, :]
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5,random_state=0) # 50% training and 50% test
```

```python
# Train Decision Tree and Visualize the Tree
clf = tree.DecisionTreeClassifier(max_depth=2,criterion='gini')
clf.fit(X=X_train, y=y_train)

fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(clf, 
                   feature_names=iris.feature_names,  
                   class_names=iris.target_names,
                   filled=True)
```

```python
y_pred = clf.predict(X_test)
# Model Accuracy
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
